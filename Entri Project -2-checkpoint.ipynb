```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Project2.csv')
df.head()
```

## Cleaning the Height column

```python
df['Height']=np.random.randint(150,181,size=len(df))
df
```

```python
df.info()
```

## Cleaning the salary data to fill the Nan values

```python
df['Salary'].fillna(df.groupby('Team')['Salary'].transform('mean'),inplace=True)
df.info()
```

```python
df.head()
```

```python
df['Salary'] = df['Salary'].round(2)
df.head()
```

## 1.How many are there in each Team and the percentage splitting with respect to the total employees. 

```python
team_counts = df['Team'].value_counts()
team_counts
```

```python
percentage_split=(team_counts/len(df))*100
percentage_split
```

## 2.Segregate the employees w.r.t different positions. 

```python
position_data=df.groupby('Position')['Name']
for position, group_df in position_data:
    print(f"\nEmployees in Position: {position}")
    print(group_df)
   
```

## 3.Find from which age group most of the employees belong to

```python
#players vs age data
Age_group_count=df['Age'].value_counts()
Age_group_count
```

```python
sns.countplot(x='Age', data=df)
plt.title('Age distribution')
plt.show()
```

```python
#The age that has the maximum number of players
Age_group_count.idxmax() 
```

## 4.Find out under which team and position, spending in terms of salary is high.

```python
#Salary distribution data based on team and position
group_salary_data = df.groupby(['Team', 'Position'])
group_salary_data.apply(lambda x: x[['Name', 'Salary']])
```

```python
#Top salaries->Team wise
highest_salary_per_group = group_salary_data.apply(lambda x: x.loc[x['Salary'].idxmax()])
print(highest_salary_per_group[['Name', 'Salary']])
```

```python
#Person recieving the highest salary
highest_salary=df['Salary'].idxmax()
person_with_highest_salary = df.loc[highest_salary]
print(person_with_highest_salary[['Name','Team','Position','Salary']])
```

## 5.Find if there is any correlation between age and salary , represent it visually.

```python
correlation_age_salary= df[['Age', 'Salary']].corr()
correlation_age_salary
```

```python
sns.regplot(x='Age', y='Salary', data=df)
plt.title('Scatter Plot of Age vs Salary')
```
